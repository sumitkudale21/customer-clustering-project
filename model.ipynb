```python
cat_feat = df[['Gender', 'Profession']]
num_feat = df[['Age', 'Annual Income ($)','Spending Score (1-100)', 'Work Experience','Family Size']]
```

```python
sns.histplot(df['Age'], kde=True)
```

```python
# Number of rows and columns for the subplots
nrows = 2
ncols = 3

# Set up the figure and axes
fig, axes = plt.subplots(nrows, ncols, figsize=(20, 15))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Iterate through each feature and plot
for i, col in enumerate(num_feat.columns):
    if num_feat[col].nunique() < 15:
        sns.countplot(data=num_feat, x=col, ax=axes[i], palette='viridis')
    else:
        sns.histplot(data=num_feat, x=col, ax=axes[i], kde=True)
    axes[i].set_title(col)

# Adjust layout to avoid overlap
plt.tight_layout()
plt.show()
```

```python
# Number of rows and columns for the subplots
nrows = 3
ncols = 2

# Set up the figure and axes
fig, axes = plt.subplots(nrows, ncols, figsize=(20, 15))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Iterate through each feature and plot
for i, col in enumerate(num_feat.columns):
    sns.boxplot(data=num_feat, x=col, ax=axes[i], palette='viridis')
    axes[i].set_title(col)

# Adjust layout to avoid overlap
plt.tight_layout()
plt.show()
```

```python
# Number of rows and columns for the subplots
nrows = 1
ncols = 2

# Set up the figure and axes
fig, axes = plt.subplots(nrows, ncols, figsize=(15, 7))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Iterate through each feature and plot
for i, col in enumerate(cat_feat.columns):
    # Get the top 10 categories
    top_10_categories = cat_feat[col].value_counts().head(10).index
    top_10_data = cat_feat[cat_feat[col].isin(top_10_categories)]
    
    sns.countplot(data=top_10_data, x=col, ax=axes[i], order=top_10_categories, palette='viridis')
    axes[i].set_title(col)
    axes[i].tick_params(axis='x', rotation=90)

# Adjust layout to avoid overlap
plt.tight_layout()
plt.show()
```

```python
# Number of rows and columns for the subplots
nrows = 2
ncols = 3

# Set up the figure and axes
fig, axes = plt.subplots(nrows, ncols, figsize=(20, 8))

# Flatten the axes array for easy iteration
axes = axes.flatten()
# Iterate through each feature and plot
for i, col in enumerate(num_feat.columns):
    sns.scatterplot(data=df, x=col, y='Spending Score (1-100)', ax=axes[i])
    axes[i].set_title(f'{col} vs. Spending Score (1-100)')

# Adjust layout to avoid overlap
plt.tight_layout()
plt.title('scatterplot based on Spending Score (1-100)')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
corr_matrix = num_feat.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df, hue='Profession', diag_kind='kde')
plt.show()
```
